## Exploratory Data Analysis (EDA)

### Question 1: Which Year recorded the highest number of air crashes?

```python
df['Year'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
df['Year'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Years with Highest Number of Air Crashes')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
plt.show()
```

# Observation:
The chart shows the years with the highest number of recorded air crashes.
the year with the highest count experienced
the greatest number of air crash incidents in the dataset.

## Question 2: Which countries experienced the most air crashes?

```python
df['Country/Region'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
df['Country/Region'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries with Most Air Crashes')
plt.xlabel('Country/Region')
plt.ylabel('Number of Crashes')
plt.show()
```

# Observation:
Th chart shows the countrie/regions with the highest number of recorded air crashes.
The country at the top of the chart experienced the most air crashes in the dataset.

### Question 3: What is the trend of air crashes over time?

```python
crashes_by_year = df['Year'].value_counts().sort_index()
crashes_by_year.plot(kind='line',figsize=(10,5))
plt.title('Trend of Air Crashes Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Air Crashes')
plt.show()
```

# Observation:
The line chart shows how the number of air crashes changed over time.
The trend helps identify periods with increasing or decreasing air crash incidents.

### Question 4: Which operators were involved in the most crashes?

```python
df['Operator'].value_counts().head(10)
```

```python
df['Operator'].value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title('Top 10 Operators Involved in Air Crashes')
plt.xlabel('Operator')
plt.ylabel('Number of Crashes')
plt.show()
```

# Observation:
The chart shows the operators that appeared most frequently in  air crash records.
Some operators had significantly more recorded crashes than others.

### Question 5: How have fatalities changed over the years?

```python
fatalities_by_year =df.groupby('Year')['Fatalities (air)'].sum()
fatalities_by_year.plot(kind='line',figsize=(10,5))
plt.title('Air Crash Fatalities Over Time')
plt.xlabel('Year')
plt.ylabel('Total Fatalities')
plt.show()
```

# Observation:
The chart shows the trend of fatalities resulting from air crashes over the years.
Some years recorded significantly higher fatalities than others,indicating periods
of more severe air crassh incidents.
